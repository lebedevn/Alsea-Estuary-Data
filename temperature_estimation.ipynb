```python
import pandas as pd
import scipy.io as sio
from scipy import signal, stats
import numpy as np
import datetime
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import re

totalMidDF = pd.read_csv('midestuary_2013_2014_processed.csv', parse_dates = ['DN'])
midDF = totalMidDF.loc[(totalMidDF["DN"] > np.datetime64('2014-01-20')) & (totalMidDF["DN"] < np.datetime64('2014-01-22'))].copy()

(slope, intercept, r, p, std_err) = stats.linregress(midDF["BottomSalCorrected"].ffill().bfill().to_numpy(), midDF["BottomTemp"].ffill().bfill().to_numpy(), alternative = "less")  
midDF["modelTemp"] = midDF["BottomSalCorrected"] * slope + intercept      
plt = px.scatter(x = midDF["BottomSalCorrected"], y = midDF["BottomTemp"])
plt.add_scattergl(x = midDF["BottomSalCorrected"], y = midDF["modelTemp"])
plt.show()


(slope, intercept, r, p, std_err) = stats.linregress(midDF["SurfaceSalCorrected"].ffill().bfill().to_numpy(), midDF["SurfaceTemp"].ffill().bfill().to_numpy(), alternative = "less")  
midDF["modelTempSurface"] = midDF["SurfaceSalCorrected"] * slope + intercept      
plt2 = px.scatter(x = midDF["SurfaceSalCorrected"], y = midDF["SurfaceTemp"])
plt2.add_scattergl(x = midDF["SurfaceSalCorrected"], y = midDF["modelTempSurface"])
plt2.show()

print(midDF.loc[(midDF["BottomSalCorrected"] >= 29.9) & (midDF["BottomSalCorrected"] <= 30.1)]["BottomTemp"].mean(skipna = True))
```
